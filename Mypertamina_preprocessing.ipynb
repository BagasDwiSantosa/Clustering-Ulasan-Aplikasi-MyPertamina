# Preprocessing

```python
df2 = pd.read_csv('/content/drive/MyDrive/dataset/04-09_mypertamina.csv')
df2.head(5)
```

```python
df2.isnull().sum()
```

```python
data2 = df2.drop([ 'appVersion','tanggal','userName', 'at'], axis=1)
data2
```

```python
data2.dropna()
```

```python
data2['content'] = data2['content'].str.lower()
data2
```

```python
def normalisasi(text):
    text = re.sub(r'[0-9]+', ' ', str(text))
    text = re.sub(r'[^\w\s]', ' ', str(text))
    text = " ".join(text.split())
    return text
```

```python
data2["comment"] = data2["content"].apply(normalisasi)
data2
```

```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory, StopWordRemover, ArrayDictionary

stopword_list = StopWordRemoverFactory().get_stop_words()
factory = StopWordRemoverFactory()
stopwords = factory.create_stop_word_remover()
```

```python
# create a new list of stopwords
new_stopwords = ['jg', 'yg','dgn']

# add both stopwords
all_stopwords = stopword_list + new_stopwords

# create ArrayDictionary
stopword_dict = ArrayDictionary(all_stopwords)

sastrawi_stopwords = StopWordRemover(stopword_dict)

def remove_stopword_with_sastrawi(text):
    text = sastrawi_stopwords.remove(text)
    return text
```

```python
data2["comment"] = data2["comment"].apply(remove_stopword_with_sastrawi)
```

```python
data2
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import swifter

# create stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()
```

```python
def text_steming(text):
    text = stemmer.stem(text)
    return text
```

```python
data2["comment2"]= data2['comment'].swifter.apply(text_steming)
```

```python
from nlp_id.lemmatizer import Lemmatizer

lemmatizer = Lemmatizer()

def text_lemmatizing(text):
    text = lemmatizer.lemmatize(text)
    return text
```

```python
data2['hasil']=data2['comment2'].swifter.apply(text_lemmatizing)
```
